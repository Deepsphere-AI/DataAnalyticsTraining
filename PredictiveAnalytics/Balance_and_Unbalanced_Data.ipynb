## Balance and Unbalanced Data

```python
import pandas as pd
```

```python
# Load the dataset (Assuming the file is named 'creditcard.csv')
df = pd.read_csv(r"C:\Users\svikr\Downloads\creditcard.csv")
```

```python
# Display the first few rows of the dataframe
df
```

```python
# Check the distribution of the target variable (Assuming it's named 'Class')
print(df['Class'].value_counts())
```

## 1st Methode

```python
from sklearn.utils import resample
```

```python
# Separate majority and minority classes
df_majority = df[df.Class == 0]
df_minority = df[df.Class == 1]
```

```python
# Downsample majority class
df_majority_downsampled = resample(df_majority,
                                  replace=False,    # sample without replacement
                                  n_samples=len(df_minority),  # to match minority class
                                  random_state=42) # reproducible results
```

```python
# Combine minority class with downsampled majority class
df_balanced = pd.concat([df_majority_downsampled, df_minority])
```

```python
# Display new class counts
print(df_balanced.Class.value_counts())
```
