# Analytics Functions

```python
# 1. LEAD() and LAG():
# Problem Statement-1: Compare a customer's total day minutes with the next customer (using LEAD).
mycursor.execute("SELECT `total day minutes`, LEAD(`total day minutes`) OVER (ORDER BY `account length`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-2: Compare total international minutes with the previous customer's (using LAG), partitioned by state.
mycursor.execute("SELECT state, `total intl minutes`, LAG(`total intl minutes`) OVER (PARTITION BY state ORDER BY `account length`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-3: Use LAG to compare the total night charge with the previous two customers, partitioned by area code and ordered by total night calls.
mycursor.execute("SELECT `area code`, `total night calls`, `total night charge`, LAG(`total night charge`, 2) OVER (PARTITION BY `area code` ORDER BY `total night calls`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# 2. NTILE():
# Problem Statement-1: Divide customers into 4 groups based on total day minutes.
mycursor.execute("SELECT `total day minutes`, NTILE(4) OVER (ORDER BY `total day minutes`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-2: Segment customers into 3 groups according to total night charge, partitioned by state.
mycursor.execute("SELECT state, `total night charge`, NTILE(3) OVER (PARTITION BY state ORDER BY `total night charge`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-3: Categorize customers into 5 tiers based on customer service calls, partitioned by state and area code.
mycursor.execute("SELECT state, `area code`, `customer service calls`, NTILE(5) OVER (PARTITION BY state, `area code` ORDER BY `customer service calls`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# 3. FIRST_VALUE() and LAST_VALUE():
# Problem Statement-1: Get the first and last total day charges in the dataset.
mycursor.execute("SELECT FIRST_VALUE(`total day charge`) OVER (ORDER BY `account length`), LAST_VALUE(`total day charge`) OVER (ORDER BY `account length`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-2: Identify the first and last total intl charge in a dataset partitioned by state.
mycursor.execute("SELECT state, FIRST_VALUE(`total intl charge`) OVER (PARTITION BY state ORDER BY `total intl calls`), LAST_VALUE(`total intl charge`) OVER (PARTITION BY state ORDER BY `total intl calls`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-3: Determine the first and last account length in a dataset ordered by total day minutes, partitioned by area code.
mycursor.execute("SELECT `area code`, `total day minutes`, FIRST_VALUE(`account length`) OVER (PARTITION BY `area code` ORDER BY `total day minutes`), LAST_VALUE(`account length`) OVER (PARTITION BY `area code` ORDER BY `total day minutes`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

# Ranking Functions

```python
# 1. ROW_NUMBER():
# Problem Statement-1: Assign a unique row number to each customer based on total day minutes.
mycursor.execute("SELECT `total day minutes`, ROW_NUMBER() OVER (ORDER BY `total day minutes`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-2: Rank customers based on the total night minutes, partitioned by state.
mycursor.execute("SELECT state, `total night minutes`, ROW_NUMBER() OVER (PARTITION BY state ORDER BY `total night minutes`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-3: Order and rank customers by total day charge, partitioned by state and area code.
mycursor.execute("SELECT state, `area code`, `total day charge`, ROW_NUMBER() OVER (PARTITION BY state, `area code` ORDER BY `total day charge`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
#  2. RANK() and DENSE_RANK():
# Problem Statement-1: Rank customers by total day minutes.
mycursor.execute("SELECT `total day minutes`, RANK() OVER (ORDER BY `total day minutes`) FROM customer_churn;")
```

```python
# Fetching result
myresult = mycursor.fetchall()
column_names = [i[0] for i in mycursor.description]
df = pd.DataFrame(myresult, columns=column_names)
df
```

```python
# Problem Statement-2: Rank customers by evening charge, allowing for ties, partitioned by area code.
mycursor.execute("SELECT `area code`, `total eve charge`, DENSE_RANK() OVER (PARTITION BY `area code` ORDER BY `total eve charge`) FROM customer_churn;")
```
